```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos para la relación Rigidez "S" del BBR vs Temperatura
temperatura = np.linspace(-10, 50, 300)  # Temperatura en °C
rigidez_S = 10 ** (4 - 0.05 * temperatura)  # Relación inversa entre Rigidez S y Temperatura
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(temperatura, rigidez_S, label="Rigidez S del BBR vs Temperatura", color="brown")
plt.xscale("linear")
plt.yscale("log")  # Escala logarítmica para rigidez S
plt.xlabel("Temperatura (°C)", fontsize=12)
plt.ylabel("Rigidez S (Pa)", fontsize=12)
plt.title("Relación entre Rigidez S del BBR y Temperatura", fontsize=14)
plt.grid(which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
