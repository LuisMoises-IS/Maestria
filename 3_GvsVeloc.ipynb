```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos para la relación G* (Módulo complejo) vs Velocidad de carga
velocidad_carga = np.logspace(-1, 2, 300)  # Velocidad de carga (de 0.1 a 100 en escala logarítmica)
G_star_carga = 10 ** (2 + 0.2 * np.log10(velocidad_carga))  # Relación directa
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(velocidad_carga, G_star_carga, label="G* vs Velocidad de carga", color="purple")
plt.xscale("log")  # Escala logarítmica para la velocidad de carga
plt.yscale("log")  # Escala logarítmica para G*
plt.xlabel("Velocidad de carga (1/s)", fontsize=12)
plt.ylabel("Módulo Complejo G* (Pa)", fontsize=12)
plt.title("Relación entre G* y Velocidad de carga", fontsize=14)
plt.grid(which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()
```

```python
# Mostrar el gráfico
plt.show()
```
