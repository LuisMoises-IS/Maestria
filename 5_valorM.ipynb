```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos para la relación Valor m vs Temperatura
temperatura = np.linspace(0, 150, 300)  # Temperatura en °C
valor_m = 0.1 + 0.005 * temperatura  # Relación directa (valor conceptual)
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(temperatura, valor_m, label="Valor m vs Temperatura", color="orange")
plt.xscale("linear")
plt.yscale("linear")  # Escala lineal para el valor m
plt.xlabel("Temperatura (°C)", fontsize=12)
plt.ylabel("Valor m", fontsize=12)
plt.title("Relación entre Valor m y Temperatura", fontsize=14)
plt.grid(which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
