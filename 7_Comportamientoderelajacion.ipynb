```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos para simular el comportamiento de relajación
tiempo = np.linspace(0, 1000, 300)  # Tiempo (segundos)
deformacion = 1 * np.exp(-0.002 * tiempo)  # Relajación exponencial (descripción general)
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(tiempo, deformacion, label="Comportamiento de Relajación", color="blue")
plt.xscale("linear")
plt.yscale("linear")  # Escala lineal para deformación
plt.xlabel("Tiempo (s)", fontsize=12)
plt.ylabel("Deformación", fontsize=12)
plt.title("Comportamiento de Relajación de un Material Viscoelástico", fontsize=14)
plt.grid(which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
