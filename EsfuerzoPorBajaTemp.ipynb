```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos para el fenómeno de esfuerzo inducido por baja temperatura
temperatura = np.linspace(0, -30, 300)  # Temperaturas bajas (°C)
esfuerzo_inducido = 10 * np.exp(0.1 * temperatura)  # Esfuerzo inducido aumenta con la baja temperatura
esfuerzo_resistente = 1 / (np.exp(-0.05 * temperatura))  # Esfuerzo resistente aumenta al bajar la temperatura
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(temperatura, esfuerzo_inducido, label="Esfuerzo Inducido", color="red")
plt.plot(temperatura, esfuerzo_resistente, label="Esfuerzo Resistente", color="blue")
plt.axvline(x=-5, linestyle="--", color="green", label="Punto de intersección")  # Punto de intersección
plt.xlabel("Temperatura (°C)", fontsize=12)
plt.ylabel("Esfuerzo (Pa)", fontsize=12)
plt.title("Fenómeno de Esfuerzo Inducido por Baja Temperatura en Asfalto", fontsize=14)
plt.grid(which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
